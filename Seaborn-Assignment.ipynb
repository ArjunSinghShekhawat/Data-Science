Q1: Five common plots that can be created using the Seaborn library and their uses are:
1. Scatter Plot: Used to visualize the relationship between two continuous variables.
2. Bar Plot: Suitable for displaying the distribution of categorical data.
3. Box Plot: Helps in visualizing the distribution and identifying outliers in numerical data.
4. Heatmap: Useful for displaying correlations in a dataset, especially in matrices.
5. Pair Plot: Enables the visualization of relationships between multiple variables in a dataset.

Q2: To load the "fmri" dataset and create a line plot for different events and regions using Seaborn:

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
plt.show()
```

Q3: To load the "titanic" dataset and create two box plots for "pclass" vs. "age" and "pclass" vs. "fare":

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
sns.boxplot(x="pclass", y="age", data=titanic)
plt.show()
sns.boxplot(x="pclass", y="fare", data=titanic)
plt.show()
```

Q4: To plot a histogram for the 'price' column in the "diamonds" dataset, with the 'cut' column as the hue:

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")
sns.histplot(data=diamonds, x="price", hue="cut", kde=True)
plt.show()
```

Q5: To create a pair plot for the "iris" dataset with the "species" column as the hue:

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
plt.show()
```

Q6: To plot a heatmap using the "flights" dataset in Seaborn:

A heatmap visualizes data in a matrix, where each cell's color represents the value. In the "flights" dataset, you can use a heatmap to show the number of passengers on flights across months and years, helping to identify seasonal patterns and trends.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")
pivot_flights = flights.pivot("month", "year", "passengers")
sns.heatmap(pivot_flights, cmap="YlGnBu")
plt.show()
```

This heatmap reveals the passenger count variations over time, making it easier to spot trends and patterns.
