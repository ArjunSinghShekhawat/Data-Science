````python
                                               #question
#Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use
#MongoDB over SQL databases?
#Q2. State and Explain the features of MongoDB.
#Q3. Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.
#Q4. Using the database and the collection created in question number 3, write a code to insert one record,
#and insert many records. Use the find() and find_one() methods to print the inserted record.
#Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
#emonstrate this.
#Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
#Q7. Explain why delete_one(), delete_many(), and drop() is used.



"""Q1. MongoDB is a NoSQL database that stores data in a flexible, JSON-like format. Non-relational databases like MongoDB are preferred for scalable, unstructured, or rapidly evolving data. MongoDB is chosen over SQL databases for its flexibility, horizontal scalability, and support for large volumes of data.

Q2. MongoDB features:
- Schema-less: No fixed structure for data.
- Horizontal scalability: Distributes data across multiple servers.
- JSON-like documents: Data stored in BSON format.
- Rich query language: Supports complex queries.
- Indexing: Provides efficient data retrieval.
- Replication and sharding: Ensures high availability and scalability.
- Aggregation framework: Allows data transformation and analysis.

Q3. Python code to connect to MongoDB, create a database, and a collection:
```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['mydb']
collection = db['mycollection']
```

Q4. Code to insert one record and many records, and print them using find() and find_one():
```python
# Insert one record
record = {"name": "John", "age": 30}
collection.insert_one(record)

# Insert many records
records = [{"name": "Alice", "age": 25}, {"name": "Bob", "age": 35}]
collection.insert_many(records)

# Print inserted records
print(collection.find_one({"name": "John"}))
for doc in collection.find():
    print(doc)
```

Q5. The find() method queries MongoDB. Example:
```python
# Find documents where age is greater than 25
result = collection.find({"age": {"$gt": 25}})
for doc in result:
    print(doc)
```

Q6. The sort() method is used to sort query results. Example:
```python
# Sort documents by age in ascending order
result = collection.find().sort("age", 1)
for doc in result:
    print(doc)
```

Q7. `delete_one()` removes one document, `delete_many()` deletes multiple documents based on a filter, 
and `drop()` deletes an entire collection. They are used for data management and cleanup in MongoDB."""
````
